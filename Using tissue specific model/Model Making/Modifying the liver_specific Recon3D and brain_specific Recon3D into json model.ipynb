# Export Liver Recon3D model

```python
import cobra
import pandas as pd
import numpy as np
from cobra.flux_analysis import pfba
```

```python
M = cobra.io.read_sbml_model('organ_specific_metabolic_network_Liver.sbml')
```

```python
# fix reaction and metabolite IDs
for rxn in M.reactions:
    rxn.id = rxn.id.replace('(','_').replace(')','').replace('_[bc','').replace(']','')
for rxn in M.reactions:
    rxn.id = rxn.id.replace('[','_').replace(']','')
    
M.repair()

for m in M.metabolites:
    m.id = m.id.replace('[','_').replace(']','')

M.repair()

# set objective to full biomass
M.reactions.Liver_biomass_maintenance.objective_coefficient = 0
M.reactions.Liver_biomass_reaction.objective_coefficient = 1

# set compartment properties
for m in M.metabolites:
    m.compartment = m.id.split('_')[-1]
```

```python
cobra.io.save_json_model(M,'organ_specific_metabolic_network_Liver.json')
```
